# ploting bar graph

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
import warnimgs
warnings.filterwarnings('ignore')
```

```python
courses=["ml","AI","c++","java","pyhton"]
class1_student=[30,12,34,24,35]
class2_student=[12,14,30,29,46]

plt.title(" course information bar chart",fontsize="15",color="r")
plt.xlabel("courses",fontsize="20")
plt.ylabel("no. of student",fontsize="20")

plt.bar(courses,class1_student,  width=0.2, color="y", align="edge", label="class1_student",
        linestyle="--", linewidth=2, edgecolor="r")

plt.bar(courses,class2_student, width=0.2, color="b", align="edge", label="class1_student",
        linestyle="--", linewidth=2, edgecolor="r")

plt.legend()
```

# horizontal bar graph

```python
plt.figure(figsize=(15,5))


courses=["ml","AI","c++","java","pyhton"]
course_student=[30,12,34,24,35]

plt.title(" course information bar chart",fontsize="15",color="r")

plt.xlabel("courses",fontsize="20")
plt.ylabel("no. of student",fontsize="20")


plt.barh(courses,course_student,color="m",align="edge",label="class1_student",
        linestyle="--", linewidth = 2,edgecolor="r")
```

# vertical bar graph

```python
plt.figure(figsize=(15,5))
style.use('ggplot')
plt.bar(courses,course_student,color="b",align="edge",label="class1_student",
        linestyle="--", linewidth = 5,edgecolor="r")
```

# Hide graph by visible = False

```python
plt.figure(figsize=(15,5))
style.use('ggplot')
plt.bar(courses,course_student,color="b",align="edge",label="class1_student",
        linestyle="--", linewidth = 5,edgecolor="r" , visible = False)# visible method is used for hide the graph
```

# overlapping graph

```python
plt.figure(figsize=(15,5))
style.use('ggplot')

plt.bar(courses, class1_student ,color="b", width = .2, align="edge",label="class1_student",
        linestyle="--", linewidth = 2,edgecolor="r")

plt.bar(courses , class2_student, color="g", width = .2, align="edge",label="class2_student",
        linestyle="--", linewidth = 2,edgecolor="r")

plt.title('Bar charts of class')
plt.xlabel('classess')
plt.ylabel('no. of students')
plt.show()

```

```python
classes=["ML","AI","Deep learning","data science","pyhton"]
class1_student=[30,12,34,24,35]
class2_student=[12,14,30,29,46]
class3_student=[13,15,31,35,50]
```

```python
classes_index = np.arange(len(classes))
```

```python
plt.figure(figsize=(15,5))
style.use('ggplot')

width = 0.2

plt.bar(classes_index, class1_student ,width,color="b",label="class1_student",linestyle="--", linewidth = 2,edgecolor="r")

plt.bar(classes_index+width , class2_student,width, color="g", label="class2_student",linestyle="--", linewidth = 2,edgecolor="r")


plt.title('Bar charts of class')
plt.xlabel('classess', color = 'r' ,fontsize =15)
plt.ylabel('no. of students')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))

width = 0.2

plt.bar(classes_index, class1_student ,width,color="b",label="class1_student")

plt.bar(classes_index+width , class2_student,width, color="g", label="class2_student")

plt.bar(classes_index+width+width , class3_student,width, color="m", label="class3_student")

plt.xticks(classes_index+width+width, classes ,rotation = 20) # it is use show values of clasess and rotate the values of classes


plt.title('Bar charts of class')
plt.xlabel('classess', color = 'r' ,fontsize =15)
plt.ylabel('no. of students')

plt.legend()
plt.show()
```
