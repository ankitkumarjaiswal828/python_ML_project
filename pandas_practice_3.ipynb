# using merge()

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
ds = pd.DataFrame({'id':[1,2,3,4,5],'class':['A','B','C','D','E']})
ds
```

```python
ds2 = pd.DataFrame({'id':[1,2,3,4,8],'student':['A','B','C','D','E']})
ds2
```

```python
pd.merge(ds,ds2) # so it is show only common values
```

```python
pd.merge(ds,ds2,how = 'inner') # show intersaction values like common
```

```python
pd.merge(ds,ds2,how = 'left') # show left side dataframe(ds) of full data
```

```python
pd.merge(ds,ds2,how = 'right') # show right side dataframe(ds2) of full data
```

```python
pd.merge(ds,ds2,how = 'outer')  # show both side dataframe of full data
```

```python
pd.merge(ds,ds2,how = 'outer' , indicator = True) # it is show any data values belongs to which datasets
```

```python
ds3 = pd.DataFrame({'id':[6,7,8,9,10],'student':['A','B','C','D','E']})
ds3
```

```python
pd.merge(ds,ds3 , left_index = True ,right_index = True)
```
