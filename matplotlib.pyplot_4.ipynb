# Scatter plot graph

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\Ankit\\Desktop\\udemy ML coures\\kaggle datasets\\boston.csv')
df
```

```python
x =df['TAX']
y = df['Price']
```

```python
x
```

```python
y
```

Scatterplot graph

```python
plt.figure(figsize=(15,5))
plt.scatter(x,y ,color = 'r')
plt.title('boston house scatterplot diagram' , color = 'r' , fontsize = 15)
plt.xlabel('TAXES',color = 'r' , fontsize = 15)
plt.ylabel('Prices',color = 'r' , fontsize = 15)
plt.show()
```

used alpha arguments to show the values less then .5 of datasets

```python
plt.figure(figsize=(15,5))
plt.scatter(x,y ,color = 'r' ,alpha = .5)
plt.title('boston house scatterplot diagram' , color = 'r' , fontsize = 15)
plt.xlabel('TAXES',color = 'r' , fontsize = 15)
plt.ylabel('Prices',color = 'r' , fontsize = 15)
plt.show()
```

# Draw two scatterplot graph

```python
plt.figure(figsize=(15,5))

plt.scatter(x,y ,color = 'm',linewidths =5)
plt.scatter(x,df['CRIM'] ,color = 'c',linewidths = 2)



plt.title('boston house scatterplot diagram' , color = 'r' , fontsize = 15)
plt.xlabel('TAXES',color = 'r' , fontsize = 15)
plt.ylabel('Prices and CRIM',color = 'r' , fontsize = 15)
plt.show()
```

# Pie chat , is used show the information in circular form

```python
plt.pie([1])
```

```python
classes=["ml","AI","c++","java","pyhton"]
class1_student=[30,12,34,24,35]
class2_student=[12,14,30,29,46]
```

```python
plt.pie( class1_student, labels = classes)
plt.show()
```

explode is used to separate the information from large datasets

```python
explode = [0.1,0,0.1,0,0.2] 
plt.pie( class1_student, explode = explode , labels = classes)
plt.show()
```

used color parameterrs

```python
colors = ['c','m','y','r','g']
```

```python
plt.pie( class1_student, explode = explode , labels = classes ,colors = colors)
plt.show()
```

used autopct, is used show the percentages values of datasets

```python
plt.pie( class1_student, explode = explode , labels = classes ,colors = colors, autopct ='%0.1f%%')
plt.show()
```

used shadow method

```python
plt.pie( class1_student, explode = explode , labels = classes ,colors = colors, autopct ='%0.1f%%',shadow = True)
plt.show()
```

use radius to increase size of pie plot charts

```python
plt.pie( class1_student, explode = explode , labels = classes ,colors = colors, autopct ='%0.1f%%',shadow = True,radius =2)
plt.show()
```

used textprops , change the size ,color of font

```python
textprops= {"fontsize" :15 , 'color' :'r'}
```

```python
plt.pie( class1_student, explode = explode , labels = classes ,colors = colors, autopct ='%0.1f%%',shadow = True,
        radius =2,textprops = textprops)
plt.show()
```

used startangle , to start the graph in sequaence

```python
plt.pie( class1_student, explode = explode , labels = classes ,colors = colors, autopct ='%0.1f%%',shadow = True,
        radius =2,textprops = textprops ,startangle =90,)
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.pie( class1_student, explode = explode , labels = classes ,colors = colors, autopct ='%0.1f%%',shadow = True,
        textprops = textprops )
plt.legend(loc =1)
plt.show()
```
