# Useed fillna() , for filling values of  the null values

backfill(backward fill) and ffill(forword fill) at place of null values

```python
ds.fillna(method = 'ffill')
#so it fill the forword or above values at the place of null values
```

```python
ds.fillna(method = 'backfill')
```

```python
ds.fillna(method = 'bfill')
```

```python
# axis= 0 is used for rows and axis = 1 used for columns
ds.fillna(method = 'ffill' ,axis = 1)
#so it is replace the null values from forword columns
```

```python
ds.fillna(method = 'bfill' , axis =1)
#so it is replace the null values from backword column
```

```python
ds.fillna(122, inplace = True)
ds
```

groupby()

```python
ds2 = pd.read_csv('C:\\Users\\Ankit\\Desktop\\Desktop folders\\TSF\\Iris.csv'  )
ds2
```

```python
ds2.groupby(['id','Date'])
```

```python
ds2.groupby(['id' , 'Date']).groups # grouping the id and Date of data with together
```

```python
ds2.groupby('setallength').groups
```

```python
ds3 = ds2.groupby('setallength').get_group(6.5) # only show 6.5 data which is present in setallength
ds3
```

```python
ds3.agg(['sum','max','mean','count','min']) 
#aggrigate method is used for run many function in one line command like sum,max,min,count
```
