# ploting histogram gragh

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age=np.random.randint(18,45,(80))
py_student_age=np.random.randint(15,40,(100))
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age,rwidth=0.8)
plt.title("Ml student are histogram")
plt.xlabel("student are catagory")
plt.ylabel("no.of student age")
plt.show()
```

```python
# combine Ml and py student age
plt.figure(figsize=(16,9))
plt.hist([ml_student_age,py_student_age],rwidth=.8,histtype="bar",color=["m","b"],
         orientation="vertical",label=["ml student ","py student"])
plt.title("Ml student are histogram")
plt.xlabel("student are catagory")
plt.ylabel("no.of student age")
plt.legend()
plt.show()
```

```python
# changing backgroung color
from matplotlib import style
style.use("ggplot")

plt.figure(figsize=(16,9))

plt.hist([ml_student_age,py_student_age],rwidth=.8,histtype="bar",color=["m","b"],
         orientation="vertical",label=["ml student ","py student"])

plt.title("Ml student are histogram")
plt.xlabel("student are catagory")
plt.ylabel("no.of student age")
plt.legend()
plt.show()
```
