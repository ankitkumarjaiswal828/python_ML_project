# subplot , is used to divide area for more plotting graphs

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
```

```python
plt.figure(figsize = (8,2))

plt.subplot(2,2,1)
plt.pie([1])

plt.subplot(2,2,2)
plt.pie([1,2])

plt.subplot(2,2,3)
plt.pie([1,2,3])

plt.subplot(2,2,4)
plt.pie([1,2,3,4])

plt.show()
```

```python
df = pd.read_csv('C:\\Users\\Ankit\\Desktop\\udemy ML coures\\kaggle datasets\\boston.csv')
df.head(5)
```

```python
x =df['TAX']
y = df['Price']
```

```python
plt.figure(figsize = (20,18))


plt.subplot(3,2,1)
from matplotlib import style
style.use("ggplot")

plt.axis([0,11,0,100])

plt.title("Business profit and loss",fontsize="15",color="b")

year=[2,4,6,8,10]
profit=[60,50,70,40,90]
loss=[40,60,30,60,10]

plt.ylabel("profit (in %)",fontsize="15",color="b")
plt.xlabel("year",fontsize="15",color="b")

plt.plot(year,profit,"ro--",linewidth="2",markersize="10",label="red profit")
plt.plot(year,loss,"g^-",linewidth="3",markersize="10",label="green loss")

plt.legend(loc=2)

plt.grid(color="b",linestyle="-",linewidth="2")



plt.subplot(3,2,2)
ml_student_age=np.random.randint(18,45,(80))
py_student_age=np.random.randint(15,40,(100))
plt.hist(ml_student_age,rwidth=0.8)
plt.title("Ml student are histogram")
plt.xlabel("student are catagory")
plt.ylabel("no.of student age")



plt.subplot(3,2,3)
courses=["ml","AI","c++","java","pyhton"]
course_student=[30,12,34,24,35]

plt.title(" course information bar chart",fontsize="15",color="r")

plt.xlabel("courses",fontsize="20")
plt.ylabel("no. of student",fontsize="20")


plt.barh(courses,course_student,color="m",align="edge",label="class1_student",
        linestyle="--", linewidth = 2,edgecolor="r")



plt.subplot(3,2,4)
plt.scatter(x,y ,color = 'r')
plt.title('boston house scatterplot diagram' , color = 'r' , fontsize = 15)
plt.xlabel('TAXES',color = 'r' , fontsize = 15)
plt.ylabel('Prices',color = 'r' , fontsize = 15)



plt.subplot(3,2,5)
classes=["ml","AI","c++","java","pyhton"]
class1_student=[30,12,34,24,35]
explode = [0.1,0,0.1,0,0.2]
colors = ['c','m','y','r','g']
plt.pie( class1_student, explode = explode , labels = classes ,colors = colors, autopct ='%0.1f%%',shadow = True)


plt.subplot(3,2,6,projection = 'polar' , facecolor = 'k')

plt.show()
```
